```python
import pandas as pd
customers = pd.DataFrame({
    'customer_id': [1, 2,3,4,5],
    'name': ['Arjun', 'Babu', 'Chandu', 'David', 'meghana'],
    'city': ['hyd', 'banglore', 'delhi', 'warangal', 'singapore']
})

print("\nCustomers Dataset")
print(customers)
transactions = pd.DataFrame({
    'transaction_id': [101, 102, 103, 104, 105, 106, 107],
    'customer_id': [1, 2, 1, 3, 4, 5, 2],
    'amount': [250, 300, 150, 400, 200, 350, 100]
})

print("\nTransactions Dataset")
print(transactions)
merged_df = pd.merge(transactions, customers, on='customer_id', how='inner')
print("\nMerged Dataset")
print(merged_df)
customer_spend = merged_df.groupby(['customer_id', 'name', 'city'])['amount'].sum().reset_index()
customer_spend.rename(columns={'amount': 'total_spend'}, inplace=True)

print("\nTotal Spend Per Customer")
print(customer_spend)
city_spend = customer_spend.groupby('city')['total_spend'].sum().reset_index()

print("\nCity-wise Spending")
print(city_spend)
top_spender = customer_spend.loc[customer_spend['total_spend'].idxmax()]

print("\nTop Spender")
print(top_spender)
print("\nInsights")
print(f"Top spender is {top_spender['name']} from {top_spender['city']} with a total spend of ${top_spender['total_spend']}.")
print("City with highest spending:", city_spend.loc[city_spend['total_spend'].idxmax(), 'city'])
```

```python
import pandas as pd
patients = pd.DataFrame({
    'patient_id': [1, 2, 3, 4, 5],
    'name': ['Anu', 'Bunty', 'Cherry', 'Dolly', 'Kinny'],
    'age': [29, 45, 34, 60, 50],
    'city': ['hyd', 'hnk', 'banglore', 'New York', 'Los Angeles']
})
print("\nPatients Dataset")
print(patients)
appointments = pd.DataFrame({
    'appointment_id': [101, 102, 103, 104, 105, 106, 107, 108],
    'patient_id': [1, 2, 1, 3, 4, 5, 2, 1],
    'doctor': ['Dr. Smith', 'Dr. Lee', 'Dr. Smith', 'Dr. Patel', 'Dr. Lee', 'Dr. Smith', 'Dr. Lee', 'Dr. Patel'],
    'date': [
        '2025-01-10', '2025-01-12', '2025-01-20', '2025-02-05',
        '2025-02-10', '2025-02-15', '2025-02-18', '2025-02-25'
    ]
})
print("\nAppointments Dataset")
print(appointments)
# 3. Merge Datasets on patient_id
merged_df = pd.merge(appointments, patients, on='patient_id', how='inner')
print("\nMerged Dataset")
print(merged_df)
# 4. Count appointments per patient
patient_visits = merged_df.groupby(['patient_id', 'name', 'age', 'city'])['appointment_id'].count().reset_index()
patient_visits.rename(columns={'appointment_id': 'visit_count'}, inplace=True)

print("\nAppointments Per Patient")
print(patient_visits)
# 5. Group by Doctor to compute workload
doctor_workload = merged_df.groupby('doctor')['appointment_id'].count().reset_index()
doctor_workload.rename(columns={'appointment_id': 'total_appointments'}, inplace=True)
print("\nDoctor Workload")
print(doctor_workload)
# 6. Find patient with maximum visits
top_patient = patient_visits.loc[patient_visits['visit_count'].idxmax()]
print("\nMost Frequent Patient")
print(top_patient)
# 7. Insights
print("\nInsights")
print(f"Most frequent patient is {top_patient['name']} with {top_patient['visit_count']} visits.")
print("Doctor with highest workload:", doctor_workload.loc[doctor_workload['total_appointments'].idxmax(), 'doctor'])
```
